```python
import matplotlib.pyplot as plt
import numpy as np
```

Line Plot

```python
x = np.linspace(0,20,1000)        #array with 1000 divisions between 0 and 20
```

```python
y = np.sin(x)+0.2*x
```

```python
plt.plot(x,y);
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('f(x)')
plt.title('PLOT');
```

```python
#Generalized XOR 
```

```python
np.random.random((4,2),)
```

```python
m = np.random.uniform(-1,1,(1500,2))
```

```python
plt.scatter(m[:,0],m[:,1])
plt.xlabel("x1")
plt.ylabel("x2");
```

```python

a1 = np.ceil(m[:,0]) 
a2 = np.ceil(m[:,1])

Y = np.logical_xor(a1,a2)
```

```python
plt.scatter(m[:,0],m[:,1],c=Y);
```
