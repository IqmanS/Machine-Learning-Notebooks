# LINEPLOTS & VIOLINPLOTS
To plot **Trends**
 - sns.lineplot() - line charts 
 - sns.violinplot() - voilin plots (combination of boxplot and kernel density estimate)

```python
insurance_data=pd.read_csv("/kaggle/input/data-for-datavis/insurance.csv")
museum_data = pd.read_csv("/kaggle/input/data-for-datavis/museum_visitors.csv",index_col="Date",parse_dates = True)
ign_data = pd.read_csv("/kaggle/input/data-for-datavis/ign_scores.csv",index_col="Platform")
candy_data = pd.read_csv("/kaggle/input/data-for-datavis/candy.csv",index_col="id")
cancer_data = pd.read_csv("/kaggle/input/data-for-datavis/cancer.csv",index_col="Id")
```

```python
plt.figure(figsize=(16,6))
plt.title("Museum Data Lineplot")
sns.lineplot(museum_data)
plt.show();
```

```python
plt.figure(figsize=(16,6))
plt.title("Museum Data Lineplot")
sns.lineplot(museum_data['Chinese American Museum'])
sns.lineplot(museum_data['Firehouse Museum']);
```

```python
plt.figure(figsize=(16,6))
plt.subplot(2,1,1)
plt.title("Museum Data Lineplot")
sns.lineplot(museum_data['Chinese American Museum'])
plt.subplot(2,1,2)
sns.lineplot(museum_data['Firehouse Museum']);
```

```python
plt.figure(figsize=(10,6))
plt.title("Museum Data Violinplot")
sns.violinplot(data=museum_data,orient="v");
```

# BAR CHARTS
To plot **Relationships**
 - sns.barplot() - Bar charts are useful for comparing quantities corresponding to different groups.
 - sns.boxplot() - Box plots are useful for analysing std and outliers of the data
 - sns.boxenplot() - Enhanced Box plot for larger datasets

```python
plt.figure(figsize=(16,8))
sns.barplot(y=ign_data.index,x=ign_data["Racing"],palette="rocket")
plt.xlabel("Rating")
plt.title("IGN Data Barplot of Racing Games by Platform");
```

```python
plt.figure(figsize=(10,5))
plt.title("IGN Game Ratings Boxplot")
sns.boxplot(ign_data,orient="h",palette="rocket");
```

```python
plt.figure(figsize=(10,5))
plt.title("IGN Game Ratings Boxenplot")
sns.boxenplot(ign_data,orient="h",palette="rocket");
```

---

# HEATMAPS
To plot **Relationships**
 - sns.heatmap - Heatmaps can be used to find color-coded patterns in tables of numbers.

```python
plt.figure(figsize=(14, 10))
plt.title("IGN Game Ratings vs Platform Heatmap")
sns.heatmap(data=ign_data, annot=True,cmap="rocket_r",vmax=9,vmin=5);
```

# SCATTER PLOTS
To plot **Relationships**
 - sns.scatterplot(): scatterplot <br>
 - sns.regplot(): add regression line to scatterplot <br>
 - sns.lmplot(): add multiple regression lines to scatterplot <br>
 - sns.swarmplot(): categorical scatter plot<br>


```python
# Scatter plot showing the relationship between 'sugarpercent' and 'winpercent'
plt.title("Candy Data Scatterplot")
sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent']);
```

```python
# Scatter plot w/ regression line showing the relationship between 'sugarpercent' and 'winpercent'
plt.title("Candy Data Scatterplot with Regression Line")
sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent']);
```

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'
plt.title("Candy Data Scatterplot based on 'Chocolate'")
sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'],hue=candy_data['chocolate'],palette="rocket")
plt.grid();
```

```python
# Color-coded scatter plot w/ regression lines
sns.lmplot(x="sugarpercent", y="winpercent", hue="chocolate", data=candy_data,palette="rocket", col="hard");
```

```python
# Scatter plot showing the relationship between 'chocolate' and 'winpercent'
plt.title("Candy Data Swarmplot")
sns.swarmplot(x = candy_data["chocolate"],y = candy_data["winpercent"],palette="rocket");
```

# HISTOGRAMS & DISTRIBUTIONS
To plot **Distributions**
 - sns.histplot(): histogram <br>
 - sns.kdeplot(): kernel density estimate (KDE) plot, a smoothed histogram <br>

```python
plt.figure(figsize=(16,8))
# Histograms for benign and maligant tumors
plt.title("Cancer Data Histplot for Benign and Maligant Tumors")
sns.histplot(cancer_data,x = "Area (mean)",hue="Diagnosis",bins=40,kde=True);
```

```python
plt.figure(figsize=(16,8))
# KDE plots for benign and malignant tumors
plt.title("Cancer Data KDE-Plot for Benign and Maligant Tumors")
sns.kdeplot(data=cancer_data,x="Radius (worst)",hue="Diagnosis",fill=True);
```

---

# MULTIPLE PLOTS & GRID PLOTS
 - sns.jointplot(kind="kde"): two-dimensional (2D) KDE plot <br>
 - pairplot(): plot pairwise relationships in a dataset <br>
 - catplot(): plotting categorical plots onto a FacetGrid.
 - sns.relplot(): plotting relations in grid


```python
plt.figure(figsize=(16,8));
sns.jointplot(data =cancer_data,x='Radius (mean)', y='Texture (mean)', kind="kde",palette="rocket_r",hue="Diagnosis",fill=True);
```

```python
plt.figure(figsize=(16,8));
sns.jointplot(data =cancer_data,x='Radius (mean)', y='Texture (mean)', kind="scatter",palette="rocket_r",hue="Diagnosis");
```

```python
sns.pairplot(insurance_data,palette="rocket",hue="sex");
```

```python
sns.catplot(data=insurance_data, x="bmi", y="region", hue="sex", kind="violin", split=True,palette="flare_r");
```

```python
sns.catplot(data=insurance_data, x="bmi", y="region", hue="sex", kind="bar",palette="flare_r");
```

```python
sns.relplot(data=candy_data, x="winpercent", y="sugarpercent", hue="chocolate",row="hard" ,col="caramel",palette="rocket");
```
