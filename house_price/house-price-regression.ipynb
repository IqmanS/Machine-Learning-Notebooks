```python
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import SGDRegressor, HuberRegressor,BayesianRidge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor, HistGradientBoostingRegressor

seed = np.random.seed(0)
cols = [i for i in house_price_data_base.columns if i!="SalePrice"]
X_train, X_test, y_train, y_test = train_test_split(house_price_data_base[cols], house_price_data_base.SalePrice, test_size=0.3,random_state=seed)
```

```python
house_price_y = house_price_data_base["SalePrice"] # Full training data
house_price_x = house_price_data_base[cols]  # Full training data

# house_price_x = X_train
# house_price_y = y_train
```

---

## 1. RandomForestRegressor

```python
rfmodel = RandomForestRegressor(n_estimators = 1000,random_state = seed,criterion="absolute_error")
rfmodel.fit(house_price_x,house_price_y)
print("MAE of Random Forest Regressor: ",mean_absolute_error(y_test,rfmodel.predict(X_test)))
```


## 2. GradientBoostingRegressor


```python
gbmodel = GradientBoostingRegressor(random_state = seed,loss="absolute_error")
gbmodel.fit(house_price_x,house_price_y)
print("MAE of Gradient Boosting Regressor: ",mean_absolute_error(y_test,gbmodel.predict(X_test)))
```


## 3. CatBoostRegressor


```python
catmodel = CatBoostRegressor(random_state = seed,loss_function="MAE",verbose=False)
catmodel.fit(house_price_x,house_price_y)
print("MAE of Category Boosting Regressor: ",mean_absolute_error(y_test,catmodel.predict(X_test)))
```


## 4. XGBRegressor


```python
xgbmodel = XGBRegressor(random_state=seed,max_depth= 6,learning_rate = 0.01,n_estimators= 5000)
xgbmodel.fit(house_price_x,house_price_y)
print("MAE of eXtreme Gradient Boosting Regressor: ",mean_absolute_error(y_test,xgbmodel.predict(X_test)))
```


## 5. LGBMRegressor


```python
lgbmodel = LGBMRegressor(random_state=seed,max_depth= 8,n_estimators= 5000)
lgbmodel.fit(house_price_x,house_price_y)
print("MAE of Light Gradient Boosted Machine Regressor: ",mean_absolute_error(y_test,lgbmodel.predict(X_test)))
```

## 6. HistGradientBoostingRegressor


```python
hgbmodel = HistGradientBoostingRegressor(random_state=seed,max_depth= 10,learning_rate = 0.15)
hgbmodel.fit(house_price_x,house_price_y)
print("MAE of Histogram-Gradient Boosted Regressor: ",mean_absolute_error(y_test,hgbmodel.predict(X_test)))
```


## 7. HuberRegressor


```python
hbmodel = HuberRegressor(max_iter = 10000)
hbmodel.fit(house_price_x,house_price_y)
print("MAE of Huber Regressor: ",mean_absolute_error(y_test,hbmodel.predict(X_test)))
```

## 8. BayesianRidge

```python
brmodel = BayesianRidge()
brmodel.fit(house_price_x,house_price_y)
print("MAE of Bayesian Ridge Regressor: ",mean_absolute_error(y_test,brmodel.predict(X_test)))
```

## 9. SGDRegressor

```python
sgdmodel = SGDRegressor()
sgdmodel.fit(house_price_x,house_price_y)
print("MAE of Stochastic Gradient Descent Regressor: ",mean_absolute_error(y_test,sgdmodel.predict(X_test)))
```

## 10. VotingRegressor


```python
vrmodel = VotingRegressor([('xgb', xgbmodel),('cat',catmodel)])
vrmodel.fit(house_price_x, house_price_y)
print("MAE of Voting Regressor:",mean_absolute_error(y_test, vrmodel.predict(X_test)))
```
